<h2>Finding Win States Function</h2>

```python
def winstate(state,player,size,seq):
    
    for j in range(0,size-seq+1):
        for i in range(0,size):
            count=0
            for k in range(0,seq):
                if state[i][k+j] == player:
                    count = count+1
            if count == seq:
                return True


    
        for i in range(0,size):
            count=0
            for k in range(0,seq):
                if state[k+j][i] == player:
                    count=count+1
            if count == seq:
                return True


    
        count=0
        for i in range(0,seq):
            if state[i+j][i+j]==player:
                count=count+1
        if count==seq:
            return True
   
        count=0
        for i in range(0,seq):
            if state[i+j][size-1-i-j]==player:
                count=count+1
        if count==seq:
            return True
        
        
         
        count=0
        for i in range(0,seq):
            if state[i+j][i] == player:
                count = count+1
        if count == seq:
            return True
  
        count =0
        for i in range(0,seq):
            if state[i][i+j] == player:
                count = count+1
        if count == seq:
            return True
  
        count=0
        for i in range(0,seq):
            if state[i][size-1-i-j] == player:
                count=count+1
        if count==seq:
            return True
  
        count=0
        for i in range(0,seq):
            if state[i+j][size-1-i] == player:
                count=count+1
        if count==seq:
            return True
        
        
    return False
   
    
```

<h2>Utility Function</h2>

```python
def evaluate(state,size,seq):
   
    if winstate(state,1,size,seq):
        evaluationval=1
    elif winstate(state,-1,size,seq):
        evaluationval=-1
    else:
        evaluationval = 0

    return evaluationval
```

<h2>Game Over Function</h2>

```python


def game_over(state,player,size,seq):
   
   
    if player==-1:
       
        if winstate(state,-1,size,seq):
            return True
        else:
            return False
    else :
       
        if winstate(state,1,size,seq):
            return True
        else:
            return False

        

```

<h2>Empty States Function</h2>

```python
def empty_pos(state):
   
    positions = []

   
               
    for i in range(0,len(state)):
        for j in range(0,len(state)):
            if state[i][j]==0:
                positions.append([i,j])
   

    return positions
```

<h1>Simple Minmax Function</h1>

```python
def minimax(state, depth, player,size,seq):
   
    if player == 1:
        max = [-1, -1, -infinity]
    else:
        max = [-1, -1, +infinity]

    if depth == 0 or game_over(state,1,size,seq) or game_over(state,-1,size,seq) :
        evaluationval = evaluate(state,size,seq)
        return [-1, -1, evaluationval]

    for pos in empty_pos(state):
   
        x, y = pos[0], pos[1]
       
        state[x][y] = player
        evaluationval = minimax(state, depth - 1, -player,size,seq)
        state[x][y] = 0
        evaluationval[0], evaluationval[1] = x, y
       

      
        if player == 1:
            max =getmax(evaluationval,max)
        else:
            max = getmin(evaluationval,max)

    return max
```

<h1>Minmax with alphabeta Pruning</h1>

```python
def alphabeta(state,depth, player,alpha,beta,size,seq):
   
    if player == 1:
        max = [-1, -1, -infinity]
        max[2]=MIN
    else:
        max = [-1, -1, +infinity]
        max[2]=MAX

    if depth == 0 or game_over(state,1,size,seq) or game_over(state,-1,size,seq) :
        evaluationval = evaluate(state,size,seq)
        return [-1, -1, evaluationval]

    for pos in empty_pos(state):
        
       
        x, y = pos[0], pos[1]
        
        state[x][y] = player
        evaluationval = alphabeta(state, depth - 1, -player,alpha,beta,size,seq)
        state[x][y] = 0
        evaluationval[0], evaluationval[1] = x, y
       

      
        if player == 1:
            max =getmax(evaluationval,max)
            
            alpha=maximum(alpha,max)
           
            if beta <= alpha:  
                break 
           
        else:
            max = getmin(evaluationval,max)
            beta = minimum(beta, max)  
            if beta <= alpha:  
                break
            
            

    return max
```

<h1>Minmax with depth limit</h1>

```python
def depthlimit(state, depth, player,size,seq,limit):
   
    if player == 1:
        max = [-1, -1, -infinity]
    else:
        max = [-1, -1, +infinity]

    if depth == limit or game_over(state,1,size,seq) or game_over(state,-1,size,seq) :
        evaluationval = evalstate(state,size,seq)
        return [-1, -1, evaluationval]

    for pos in empty_pos(state):
    
        x, y =pos[0],pos[1]
        state[x][y] = player
        evaluationval = depthlimit(state, depth - 1, -player,size,seq,limit)
        state[x][y] = 0
        evaluationval[0], evaluationval[1] = x, y
       

      
        if player == 1:
            max =getmax(evaluationval,max)
        else:
            max = getmin(evaluationval,max)

    return max

```

```python
def depthlimitalphabeta(state,depth, player,alpha,beta,size,seq,limit):
   
    if player == 1:
        max = [-1, -1, -infinity]
        max[2]=MIN
    else:
        max = [-1, -1, +infinity]
        max[2]=MAX

    if depth == limit or game_over(state,1,size,seq) or game_over(state,-1,size,seq) :
        evaluationval = evalstate(state,size,seq)
        return [-1, -1, evaluationval]

    for pos in empty_pos(state):
   
        x, y = pos[0], pos[1]
       
        state[x][y] = player
        evaluationval = depthlimitalphabeta(state, depth - 1, -player,alpha,beta,size,seq,limit)
        state[x][y] = 0
        evaluationval[0], evaluationval[1] = x, y
       

      
        if player == 1:
            max =getmax(evaluationval,max)
            
            alpha=maximum(alpha,max)
           
            if beta <= alpha:  
                break 
           
        else:
            max = getmin(evaluationval,max)
            beta = minimum(beta, max)  
            if beta <= alpha:  
                break
            
            

    return max



```

<h1>Minmax with Heuristic Function</h1>

```python
def heuristic(state, depth, player,size,seq):
   
    if player == 1:
        max = [-1, -1, -infinity]
    else:
        max = [-1, -1, +infinity]

    if depth == 0 or game_over(state,1,size,seq) or game_over(state,-1,size,seq) :
        evaluationval = evalstate(state,size,seq)
        return [-1, -1, evaluationval]
    max1=-infinity
    for i in range(0,len(state)):
        for j in range(0,len(state)):
            if state[i][j]==0:
               
                state[i][j]=player
                evaluationval2=evalstate(state,size,seq)
                state[i][j]=0
                if max1<evaluationval2:
                    max1=evaluationval2
                   
                    x=i
                    y=j
                
                
       
    state[x][y] =player 
    evaluationval = heuristic(state, depth - 1, -player,size,seq)
    state[x][y] = 0
    evaluationval[0], evaluationval[1] = x, y       

      
    if player == 1:
            max =getmax(evaluationval,max)
    else:
            max = getmin(evaluationval,max)

    return max
```
